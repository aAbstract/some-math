```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pin_to_hz(pin_idx: int, fs: float, N: int):
    return (fs * pin_idx) / N
```

```python
FS = 1000  # Hz
T = 1  # s
t = np.linspace(0, T, T*FS, endpoint=False)
signal = 5 * np.cos(2 * np.pi * 100 * t - np.pi / 6)

plt.plot(t, signal)
```

```python
hz_axis = np.array([pin_to_hz(pin_idx=p, fs=FS, N=T * FS) for p in range(0, T * FS)])

sig_fft = np.fft.fft(signal)
mag_fft = np.abs(sig_fft) / (FS * T)

plt.grid()
plt.xticks(np.arange(0, 1001, 100))
plt.plot(hz_axis, mag_fft)
```
