```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
FS = 1000
T1 = 0
T2 = 5
t = np.linspace(T1, T2, (T2 - T1) * FS, endpoint=False)

y_list = [
    np.ones(shape=t.shape),
    np.ones(shape=t.shape),
    0.5 * np.ones(shape=t.shape),
    -1 * np.ones(shape=t.shape),
    np.exp(-t / 5),
    np.exp(-t),
    np.sin(2 * np.pi * t),
    np.ones(shape=t.shape)
]

enrg_list = [(1 / FS) * np.sum(y**2) for y in y_list]
corr_list = []
for idx, y in enumerate(y_list):
    enrg_f = 1 / np.sqrt(enrg_list[0] * enrg_list[idx])
    corr = enrg_f * (1 / FS) * np.sum(y_list[0] * y_list[idx])
    corr_list.append(np.round(corr, 3))

fig, axes = plt.subplots(2, 4, figsize=(18, 6))
for idx, ax in enumerate(axes.flat):
    if idx == 0:
        ax.set_title('Base Signal')
    else:
        ax.set_title(f"Target Signal {idx}, Corr = {corr_list[idx]}")

    ax.set_ylim(-1.5, 1.5)
    ax.plot(t, y_list[idx])
```
